```python
import requests

r = requests.get('https://www.dcard.tw/_api/forums/pet/posts?popular=true')
response = r.text
```

```python
import json

data = json.loads(response)

# 這個 API 一次會回傳幾筆資料
print('這個 API 一次會回傳',len(data),'筆資料')
```

```python
# 取出每一筆資料的「標題」、「貼文時間」、「留言人數」、「按讚人數」
for d in data:
    print('標題：', d['title'])
    print('貼文時間：', d['createdAt'])
    print('留言人數：', d['commentCount'])
    print('按讚人數：', d['likeCount'])
    print('')
```

```python
# 計算熱門/非熱門文章的「平均留言人數」與「平均按讚人數」
comment_sum = 0
like_sum = 0

for d in data:
    comment_sum = comment_sum + d['commentCount']
    like_sum = like_sum + d['likeCount']

print('熱門文章')
print('平均留言人數：', round(comment_sum/len(data), 2))
print('平均按讚人數：', round(like_sum/len(data), 2))
print('')

r = requests.get('https://www.dcard.tw/_api/forums/pet/posts?popular=false')
data = json.loads(r.text)

comment_sum = 0
like_sum = 0

for d in data:
    comment_sum = comment_sum + d['commentCount']
    like_sum = like_sum + d['likeCount']

print('非熱門文章')
print('平均留言人數：', round(comment_sum/len(data), 2))
print('平均按讚人數：', round(like_sum/len(data), 2))
```
